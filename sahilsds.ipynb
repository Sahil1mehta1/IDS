```python
manhattan_df = sahil[sahil['neighbourhood_group'] == 'Manhattan']#filter for neighbourhood group.
brooklyn_df = sahil[sahil['neighbourhood_group'] == 'Brooklyn']
manhattan_room_counts = manhattan_df['room_type'].value_counts()#room type counts.
brooklyn_room_counts = brooklyn_df['room_type'].value_counts()
plt.figure(figsize=(6, 6))#plot room type 
plt.pie(manhattan_room_counts, labels=manhattan_room_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Room Type Distribution in Manhattan')
plt.show()
plt.figure(figsize=(6, 6))#plot room type
plt.pie(brooklyn_room_counts, labels=brooklyn_room_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Room Type Distribution in Brooklyn')
plt.show()
```

```python
#multivariate analysis with pairplot.
selected_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']
clean_data = sahil[selected_columns + ['neighbourhood_group']].dropna()
sns.pairplot(clean_data, vars=selected_columns, hue='neighbourhood_group')
plt.suptitle("Pairplot of Selected Features Colored by Neighbourhood Group", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))# multivariate scatter plot.
sns.scatterplot(x='price',y='minimum_nights',hue='neighbourhood_group',size='availability_365',style='room_type',data=sahil,palette='Set1',sizes=(50, 250),alpha=0.7)
plt.title('Multivariate Scatter Plot: Price vs Minimum Nights')
plt.xlabel('Price ($)')
plt.ylabel('Minimum Nights')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
corr_matrix = sahil.corr(numeric_only=True)#correlation heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Airbnb NYC)')
plt.show()
```
